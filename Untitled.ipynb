## Data Exploration



```python
import numpy as np 
import pandas as pd
import json  # Import the json module
from pandas import json_normalize
from sklearn.model_selection import train_test_split
import plotly.express as px
import matplotlib.pyplot as plt
from pandas import get_dummies
```

## Leser inn data


```python
df_hospital=pd.read_csv('raw_data/hospital.csv')
df_hospital.head()
```

```python
df_demographic=pd.read_csv('raw_data/demographic.csv')
```

#### gjør om alt til dummy variabler

```python
df_demographic=get_dummies(df_demographic, columns=['kjønn', 'etnisitet'])
#1 representerer male og 0 representerer female. 
df_demographic['kjønn'] = df_demographic['kjønn_male']
df_demographic.drop(columns=['kjønn_female', 'kjønn_male'], inplace=True)

income_mapping= {
    'under $11k': 0,
    '$11k - $25k': 1,
    '$25k - $50k': 2,
    '>$50k': 3,
}
df_demographic['inntekt']=df_demographic['inntekt'].map(income_mapping)
df_demographic=df_demographic.replace({True:1, False:0})
#df_demographic.head()
#df_demographic.info()
```

```python
df_physiological=pd.read_table('raw_data/physiological.txt')
#df_physiological.head()
#df_physiological.info()
```

```python
json_df=pd.read_json('raw_data/severity.json')
#json_df.head()

```

## Optimalisere Json fil og slår sammen tabeller

```python
#resonnomant
#explode ikke sykdomskategori-id, har ikke lister har kun et element. Likt som sykdomskategori,
#tar med de kun med liste og de som er viktige
#Optimaliserer, kan da slåes sammen med andre kategoriene. 
```

```python
df_severity=json_df.iloc[:,0:-1].explode(list(json_df.columns[2:-1]))
df_severity.reset_index(drop=True,inplace=True)
df_severity=df_severity.sort_values(by=['pasient_id'], ignore_index=True)

```

```python
df_severity= pd.get_dummies(df_severity, columns=['sykdomskategori_id', 'sykdomskategori', 'sykdom_underkategori'])
#df_severity.head()
```

```python
#slår sammen og står med endelige dataFramen med alle data som vi vil jobbe etter. 
df=df_demographic.merge(df_hospital,on='pasient_id',how='outer')\
    .merge(df_physiological,on='pasient_id',how='outer')\
    .merge(df_severity,on='pasient_id',how='outer')
#df
```

```python
#fjerne der opphold er ugyldig, aka minus. 
df = df.loc[df['oppholdslengde'] >= 0].reset_index(drop=True)
#df
```

```python
#fjerne duplikater
df=df.drop_duplicates(subset=['pasient_id']).reset_index(drop=True)
#df.info()
```

## Mapper Kreft verdier til dummyvariabler

```python
df= pd.get_dummies(df, columns=['kreft'])
df.drop(columns=['kreft_no'],inplace=True)
df['kreft_yes'] = df[['kreft_metastatic', 'kreft_yes']].max(axis=1)
df.rename(columns={'kreft_yes': 'kreft'}, inplace=True)
df=df.replace({True:1, False:0})

```

## Mapper dnr_status

```python
dnr_mapping = {
    'dnr ved innleggelse': 2,
    'dnr før innleggelse': 1,
    'ingen dnr': 0,
}

# Map DNR-status verdiene til numeriske verdier
df['dnr_status'] = df['dnr_status'].map(dnr_mapping)
df['dnr_status'] = df['dnr_status'].astype(float)
```

## slår sammen adl

```python
df['adl'] = df[['adl_pasient', 'adl_stedfortreder']].mean(axis=1)
df.drop(columns=['adl_pasient', 'adl_stedfortreder'], inplace=True)
```
